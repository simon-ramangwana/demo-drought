```python
# ============================================================
# Cell 1 — CMIP6 Historical Configuration
# ============================================================

from pathlib import Path

START_YEAR = 1981
END_YEAR = 2014

YEARS = [str(y) for y in range(START_YEAR, END_YEAR + 1)]

CLIMATE_MODEL = "MPI-ESM1-2-LR"

VARIABLES = {
    "tas": "near_surface_air_temperature",
    "huss": "near_surface_specific_humidity",
    "pr": "precipitation",
    "mrsos": "moisture_in_upper_portion_of_soil_column",
    "rsds": "surface_downwelling_shortwave_radiation"
}

HISTORICAL_ROOT = Path(
    r"C:\Projects\Infer RozviDrought\data\historical\cmip6"
)

HISTORICAL_ROOT.mkdir(parents=True, exist_ok=True)

print("Historical CMIP6 root:")
print(HISTORICAL_ROOT)
```

```python
# ============================================================
# Cell 2 — Create Historical CMIP6 Folder Structure
# ============================================================

created_paths = []

for var in VARIABLES.keys():
    path = HISTORICAL_ROOT / var
    path.mkdir(parents=True, exist_ok=True)
    created_paths.append(path)

print("Folders created / verified:")
for p in created_paths:
    print(p)
```

```python
# ============================================================
# Cell 3 — Download CMIP6 Historical Data
# ============================================================

import cdsapi

c = cdsapi.Client()

downloaded = []
skipped = []
failed = []

for var, cds_name in VARIABLES.items():

    target_file = (
        HISTORICAL_ROOT
        / var
        / f"{var}_{CLIMATE_MODEL}_historical_{START_YEAR}_{END_YEAR}.nc"
    )

    if target_file.exists():
        print("Skipping existing:", target_file.name)
        skipped.append((var, "already_exists"))
        continue

    print("\nDownloading historical:", var)

    try:
        c.retrieve(
            "projections-cmip6",
            {
                "temporal_resolution": "monthly",
                "experiment": "historical",
                "variable": cds_name,
                "model": CLIMATE_MODEL,
                "year": YEARS,
                "format": "netcdf",
            },
            str(target_file),
        )

        print("Success:", var)
        downloaded.append((var, str(target_file)))

    except Exception as e:
        print("Failed:", var)
        print(e)
        failed.append((var, str(e)))

print("\n===================================")
print("Downloaded:")
for item in downloaded:
    print(item)

print("\n===================================")
print("Skipped:")
for item in skipped:
    print(item)

print("\n===================================")
print("Failed:")
for item in failed:
    print(item)
```

```python
# ============================================================
# Cell 4 — Unzip Historical CMIP6 Files
# ============================================================

import zipfile

unzipped = []

for var in VARIABLES.keys():

    var_dir = HISTORICAL_ROOT / var

    for file in var_dir.glob("*.nc"):

        with open(file, "rb") as f:
            signature = f.read(2)

        # Check if file is actually a zip archive
        if signature == b"PK":

            print("Unzipping:", file.name)

            with zipfile.ZipFile(file, "r") as z:
                z.extractall(var_dir)

            unzipped.append(file.name)

print("\n===================================")
print("Files unzipped:", len(unzipped))
```

```python
# ============================================================
# Cell 5 — Verify Historical CMIP6 Files
# ============================================================

import xarray as xr

print("===================================")
print("HISTORICAL FILE VALIDATION")
print("===================================")

results = []

for var in VARIABLES.keys():

    var_dir = HISTORICAL_ROOT / var

    files = list(var_dir.glob("*.nc"))

    if not files:

        results.append({
            "variable": var,
            "status": "missing_file"
        })

        continue

    file = files[0]

    try:

        ds = xr.open_dataset(file)

        results.append({
            "variable": var,
            "status": "ok",
            "file_name": file.name,
            "time_steps": ds.sizes["time"],
            "lat_size": ds.sizes["lat"],
            "lon_size": ds.sizes["lon"],
            "time_start": str(ds.time.values[0]),
            "time_end": str(ds.time.values[-1])
        })

        ds.close()

    except Exception as e:

        results.append({
            "variable": var,
            "status": "failed_to_open",
            "file_name": file.name,
            "error": str(e)
        })

for r in results:
    print(r)
```

```python
# ============================================================
# Cell — Save Historical CMIP6 Download Manifest
# ============================================================

import json
from datetime import datetime, UTC

MANIFEST_ROOT = Path(
    r"C:\Projects\Infer RozviDrought\data\artifacts\manifests"
)

MANIFEST_ROOT.mkdir(parents=True, exist_ok=True)

timestamp = datetime.now(UTC).strftime("%Y%m%dT%H%M%SZ")

records = []

for var in VARIABLES.keys():

    var_dir = HISTORICAL_ROOT / var

    files = list(var_dir.glob("*.nc"))

    if not files:

        records.append({
            "variable": var,
            "status": "missing"
        })

        continue

    file = files[0]

    try:

        import xarray as xr

        ds = xr.open_dataset(file)

        record = {
            "variable": var,
            "file_name": file.name,
            "path": str(file),
            "time_steps": int(ds.sizes["time"]),
            "lat_size": int(ds.sizes["lat"]),
            "lon_size": int(ds.sizes["lon"]),
            "time_start": str(ds.time.values[0]),
            "time_end": str(ds.time.values[-1])
        }

        ds.close()

        records.append(record)

    except Exception as e:

        records.append({
            "variable": var,
            "file_name": file.name,
            "status": "failed_to_read",
            "error": str(e)
        })


manifest = {
    "dataset": "cmip6_historical",
    "model": CLIMATE_MODEL,
    "period": {
        "start_year": START_YEAR,
        "end_year": END_YEAR
    },
    "variables": list(VARIABLES.keys()),
    "created_utc": timestamp,
    "records": records
}

manifest_file = (
    MANIFEST_ROOT
    / f"historical_cmip6_manifest_{CLIMATE_MODEL}_{START_YEAR}_{END_YEAR}_{timestamp}.json"
)

with open(manifest_file, "w") as f:
    json.dump(manifest, f, indent=4)

print("===================================")
print("Historical manifest saved:")
print(manifest_file)
print("Total records:", len(records))
```
