```python
print# ---------------------------------------------------------------------
# FIND MISSING HISTORICAL TWS MONTHS
# ---------------------------------------------------------------------

from pathlib import Path

print("=" * 60)
print("MISSING HISTORICAL TWS MONTHS")
print("=" * 60)

PROJECT_ROOT = Path(r"C:\Projects\Infer RozviDrought")

TWS_OUT_DIR = (
    PROJECT_ROOT
    / "rasters"
    / "corrected"
    / "proxies"
    / "tws"
    / "historical"
)

EXPECTED_HIST_START = "198109"
EXPECTED_HIST_END = "202409"


def extract_yyyymm(path):
    stem = path.stem
    for token in stem.split("_"):
        if token.isdigit() and len(token) == 6:
            return token
    return None


def month_range(start, end):
    months = []
    y = int(start[:4])
    m = int(start[4:])

    while True:
        ym = f"{y:04d}{m:02d}"
        months.append(ym)

        if ym == end:
            break

        m += 1
        if m > 12:
            m = 1
            y += 1

    return months


files = sorted(TWS_OUT_DIR.glob("*.tif"))
months_found = sorted(
    extract_yyyymm(fp)
    for fp in files
    if extract_yyyymm(fp) is not None
)

expected_months = month_range(EXPECTED_HIST_START, EXPECTED_HIST_END)
missing = sorted(set(expected_months) - set(months_found))

print(f"Months found   : {len(months_found)}")
print(f"Expected       : {len(expected_months)}")
print(f"Missing count  : {len(missing)}")
print("\nMissing months:")
print(missing)
```
