```python
# ============================================================
# CELL 5 — REUSABLE ALIGNMENT FUNCTION WITH NAN HANDLING
# ============================================================

print("============================================================")
print("BUILDING REUSABLE ALIGNMENT FUNCTION")
print("============================================================")

from pathlib import Path

import numpy as np
import rasterio
from rasterio.enums import Resampling
from rasterio.transform import from_origin
from rasterio.warp import reproject
import rozvidrought_datasets.grid.spec as grid_spec


# ------------------------------------------------------------
# Authoritative template
# ------------------------------------------------------------

GRID_WIDTH = grid_spec.WIDTH
GRID_HEIGHT = grid_spec.HEIGHT
GRID_LEFT = grid_spec.LEFT
GRID_TOP = grid_spec.TOP
GRID_RIGHT = grid_spec.RIGHT
GRID_BOTTOM = grid_spec.BOTTOM
GRID_CRS = grid_spec.CRS
GRID_PIXEL_SIZE_X = grid_spec.PIXEL_SIZE_X
GRID_PIXEL_SIZE_Y = grid_spec.PIXEL_SIZE_Y

GRID_TRANSFORM = from_origin(
    west=GRID_LEFT,
    north=GRID_TOP,
    xsize=abs(GRID_PIXEL_SIZE_X),
    ysize=abs(GRID_PIXEL_SIZE_Y),
)

DEFAULT_NODATA = -9999.0


# ------------------------------------------------------------
# Helper
# ------------------------------------------------------------

def summarize_array(arr, nodata=DEFAULT_NODATA):
    finite_mask = np.isfinite(arr) & (arr != nodata)

    summary = {
        "valid_pixels": int(finite_mask.sum()),
        "nodata_pixels": int((arr == nodata).sum()),
        "nan_pixels": int(np.isnan(arr).sum()),
    }

    if finite_mask.any():
        summary["min"] = float(arr[finite_mask].min())
        summary["max"] = float(arr[finite_mask].max())
        summary["mean"] = float(arr[finite_mask].mean())
    else:
        summary["min"] = None
        summary["max"] = None
        summary["mean"] = None

    return summary


# ------------------------------------------------------------
# Main aligner
# ------------------------------------------------------------

def align_raster_to_authoritative_grid(
    src_path,
    dst_path,
    resampling=Resampling.bilinear,
    dst_nodata=DEFAULT_NODATA,
    dtype="float32",
):
    src_path = Path(src_path)
    dst_path = Path(dst_path)
    dst_path.parent.mkdir(parents=True, exist_ok=True)

    if not src_path.exists():
        raise FileNotFoundError(f"Source raster not found: {src_path}")

    dst_array = np.full((GRID_HEIGHT, GRID_WIDTH), dst_nodata, dtype=np.float32)

    with rasterio.open(src_path) as src:
        src_array = src.read(1).astype(np.float32)

        # convert NaNs in source to nodata before reprojection
        src_array = np.where(np.isfinite(src_array), src_array, dst_nodata)

        src_nodata = src.nodata if src.nodata is not None else dst_nodata

        reproject(
            source=src_array,
            destination=dst_array,
            src_transform=src.transform,
            src_crs=src.crs,
            src_nodata=src_nodata,
            dst_transform=GRID_TRANSFORM,
            dst_crs=GRID_CRS,
            dst_nodata=dst_nodata,
            resampling=resampling,
        )

        dst_profile = src.profile.copy()
        dst_profile.update(
            driver="GTiff",
            height=GRID_HEIGHT,
            width=GRID_WIDTH,
            count=1,
            dtype=dtype,
            crs=GRID_CRS,
            transform=GRID_TRANSFORM,
            nodata=dst_nodata,
            compress="deflate",
        )

    with rasterio.open(dst_path, "w", **dst_profile) as dst:
        dst.write(dst_array.astype(dtype), 1)

    with rasterio.open(dst_path) as check:
        arr = check.read(1)
        summary = summarize_array(arr, nodata=dst_nodata)

        result = {
            "src_path": str(src_path),
            "dst_path": str(dst_path),
            "width": check.width,
            "height": check.height,
            "crs": str(check.crs),
            "bounds": tuple(check.bounds),
            "transform": check.transform,
            "nodata": check.nodata,
            **summary,
        }

    return result


print("Reusable function ready:")
print(" - summarize_array(arr, nodata=-9999.0)")
print(" - align_raster_to_authoritative_grid(src_path, dst_path, ...)")
print("============================================================")
```

```python
# ============================================================
# CELL 12 — GENERATE MISSING HISTORICAL MONTHS FROM NETCDF
# FIXED: handles valid_time / time dimensions correctly
# ============================================================

print("============================================================")
print("GENERATE MISSING HISTORICAL MONTHS FROM NETCDF")
print("============================================================")

from pathlib import Path
import re

import numpy as np
import xarray as xr
import rasterio
from rasterio.transform import from_bounds
from rasterio.enums import Resampling
from rasterio.warp import reproject

RAW_BASE = Path(r"C:\Projects\Infer RozviDrought\data\simulated\era5_land")
ALIGNED_BASE = Path(r"C:\Projects\Infer RozviDrought\data\aligned_grid\historical")

VARIABLES = ["t2m", "d2m", "pet"]
DST_NODATA = -9999.0


# ------------------------------------------------------------
# Helpers
# ------------------------------------------------------------

def extract_yyyymm(path):
    m = re.search(r"(\d{6})", path.name)
    return m.group(1) if m else None


def pick_data_var(ds, expected_name):
    candidates = list(ds.data_vars)

    if expected_name in ds.data_vars:
        return expected_name

    lower_map = {k.lower(): k for k in ds.data_vars}
    if expected_name.lower() in lower_map:
        return lower_map[expected_name.lower()]

    for name in candidates:
        if expected_name.lower() in name.lower():
            return name

    if len(candidates) == 1:
        return candidates[0]

    raise ValueError(
        f"Could not determine data variable for '{expected_name}'. "
        f"Available: {candidates}"
    )


def pick_coord_name(ds, options):
    ds_names = set(ds.coords) | set(ds.variables)
    for name in options:
        if name in ds_names:
            return name
    return None


def reduce_to_2d(da):
    # drop/select all non-spatial dims safely
    spatial_candidates = {"lat", "latitude", "y", "lon", "longitude", "x"}

    for dim in list(da.dims):
        if dim.lower() not in spatial_candidates:
            if da.sizes[dim] == 1:
                da = da.isel({dim: 0})
            else:
                da = da.isel({dim: 0})

    return da


def load_nc_as_2d_array(nc_path, expected_var):
    with xr.open_dataset(nc_path) as ds:
        data_var = pick_data_var(ds, expected_var)
        da = ds[data_var]

        # crucial fix: remove valid_time / time / any other non-spatial dims first
        da = reduce_to_2d(da)

        lat_name = pick_coord_name(ds, ["lat", "latitude", "y"])
        lon_name = pick_coord_name(ds, ["lon", "longitude", "x"])

        if lat_name is None or lon_name is None:
            raise ValueError(
                f"Could not find lat/lon coordinates in {nc_path.name}. "
                f"Coords: {list(ds.coords)} | Vars: {list(ds.variables)}"
            )

        # use ... so xarray does not complain if dimension order differs
        da = da.transpose(lat_name, lon_name, ...)

        # after transpose/reduction, must be 2D
        if da.ndim != 2:
            raise ValueError(
                f"Expected 2D array after reduction, got shape {da.shape} "
                f"with dims {da.dims} in {nc_path.name}"
            )

        arr = da.values.astype(np.float32)

        lats = ds[lat_name].values
        lons = ds[lon_name].values

    # ensure north-up orientation
    if lats[0] < lats[-1]:
        arr = np.flipud(arr)
        lats = lats[::-1]

    left = float(np.min(lons))
    right = float(np.max(lons))
    bottom = float(np.min(lats))
    top = float(np.max(lats))

    if len(lons) < 2 or len(lats) < 2:
        raise ValueError(f"Not enough coordinate values in {nc_path.name}")

    xres = float(np.abs(lons[1] - lons[0]))
    yres = float(np.abs(lats[1] - lats[0]))

    # convert center coords to outer bounds
    src_bounds = (
        left - xres / 2.0,
        bottom - yres / 2.0,
        right + xres / 2.0,
        top + yres / 2.0,
    )

    src_transform = from_bounds(
        *src_bounds,
        width=arr.shape[1],
        height=arr.shape[0],
    )

    return arr, src_transform, data_var


def align_nc_to_authoritative_grid(
    nc_path,
    dst_path,
    expected_var,
    resampling=Resampling.bilinear,
):
    nc_path = Path(nc_path)
    dst_path = Path(dst_path)
    dst_path.parent.mkdir(parents=True, exist_ok=True)

    src_arr, src_transform, data_var = load_nc_as_2d_array(nc_path, expected_var)
    src_arr = np.where(np.isfinite(src_arr), src_arr, DST_NODATA)

    dst_arr = np.full((GRID_HEIGHT, GRID_WIDTH), DST_NODATA, dtype=np.float32)

    reproject(
        source=src_arr,
        destination=dst_arr,
        src_transform=src_transform,
        src_crs=GRID_CRS,
        src_nodata=DST_NODATA,
        dst_transform=GRID_TRANSFORM,
        dst_crs=GRID_CRS,
        dst_nodata=DST_NODATA,
        resampling=resampling,
    )

    profile = {
        "driver": "GTiff",
        "height": GRID_HEIGHT,
        "width": GRID_WIDTH,
        "count": 1,
        "dtype": "float32",
        "crs": GRID_CRS,
        "transform": GRID_TRANSFORM,
        "nodata": DST_NODATA,
        "compress": "deflate",
    }

    with rasterio.open(dst_path, "w", **profile) as dst:
        dst.write(dst_arr, 1)

    summary = summarize_array(dst_arr, nodata=DST_NODATA)

    return {
        "src_path": str(nc_path),
        "dst_path": str(dst_path),
        "data_var": data_var,
        "valid_pixels": summary["valid_pixels"],
        "nodata_pixels": summary["nodata_pixels"],
        "nan_pixels": summary["nan_pixels"],
        "min": summary["min"],
        "max": summary["max"],
        "mean": summary["mean"],
    }


# ------------------------------------------------------------
# Generate only missing aligned months
# ------------------------------------------------------------

all_results = {}
all_failures = {}

for var in VARIABLES:
    raw_dir = RAW_BASE / var
    aligned_dir = ALIGNED_BASE / var
    aligned_dir.mkdir(parents=True, exist_ok=True)

    nc_files = sorted(raw_dir.glob(f"{var}_*.nc"))
    tif_files = sorted(aligned_dir.glob(f"{var}_*.tif"))

    nc_map = {extract_yyyymm(p): p for p in nc_files if extract_yyyymm(p)}
    tif_months = {extract_yyyymm(p) for p in tif_files if extract_yyyymm(p)}

    missing_months = sorted(set(nc_map) - tif_months)

    print("\n------------------------------------------------------------")
    print("VARIABLE       :", var)
    print("NC FILES       :", len(nc_files))
    print("ALIGNED TIFS   :", len(tif_files))
    print("TO GENERATE    :", len(missing_months))

    results = []
    failures = []

    for i, yyyymm in enumerate(missing_months, start=1):
        nc_path = nc_map[yyyymm]
        dst_path = aligned_dir / f"{var}_{yyyymm}.tif"

        try:
            result = align_nc_to_authoritative_grid(
                nc_path=nc_path,
                dst_path=dst_path,
                expected_var=var,
                resampling=Resampling.bilinear,
            )
            result["yyyymm"] = yyyymm
            result["status"] = "ok"
            results.append(result)

            if i <= 3 or i == len(missing_months):
                print(
                    f"[{i}/{len(missing_months)}] OK : {yyyymm} -> {dst_path.name} "
                    f"(min={result['min']}, max={result['max']}, mean={result['mean']})"
                )

        except Exception as e:
            failures.append({
                "yyyymm": yyyymm,
                "src_path": str(nc_path),
                "dst_path": str(dst_path),
                "status": "failed",
                "error": str(e),
            })
            print(f"[{i}/{len(missing_months)}] FAILED : {yyyymm} | {e}")

    all_results[var] = results
    all_failures[var] = failures

    print("SUCCESS        :", len(results))
    print("FAILURES       :", len(failures))

print("\n============================================================")
print("NETCDF BACKFILL COMPLETE")
print("============================================================")
```

```python
# ============================================================
# CELL 19 — CHECK PROXY GRID MATCH AGAINST AUTHORITATIVE GRID
# ============================================================

print("============================================================")
print("CHECK PROXY GRID MATCH AGAINST AUTHORITATIVE GRID")
print("============================================================")

from pathlib import Path
import rasterio
import rozvidrought_datasets.grid.spec as grid_spec

BASE_DIR = Path(r"C:\Projects\Infer RozviDrought\rasters\corrected\proxies")

CHECK_TARGETS = [
    ("ndvi", "historical"),
    ("ndvi", "ssp245"),
    ("tws", "historical"),
    ("tws", "ssp245"),
]

pkg_width = grid_spec.WIDTH
pkg_height = grid_spec.HEIGHT
pkg_left = grid_spec.LEFT
pkg_right = grid_spec.RIGHT
pkg_top = grid_spec.TOP
pkg_bottom = grid_spec.BOTTOM
pkg_crs = str(grid_spec.CRS)
pkg_px = grid_spec.PIXEL_SIZE_X
pkg_py = grid_spec.PIXEL_SIZE_Y

for var, scenario in CHECK_TARGETS:
    src_dir = BASE_DIR / var / scenario
    files = sorted(src_dir.glob("*.tif"))

    print("\n============================================================")
    print(f"VARIABLE : {var}")
    print(f"SCENARIO : {scenario}")
    print("============================================================")

    if not files:
        print("No files found")
        continue

    sample = files[0]

    with rasterio.open(sample) as src:
        bounds = src.bounds
        px = src.transform.a
        py = src.transform.e

        same_grid = (
            src.width == pkg_width
            and src.height == pkg_height
            and str(src.crs) == pkg_crs
            and abs(bounds.left - pkg_left) < 1e-9
            and abs(bounds.right - pkg_right) < 1e-9
            and abs(bounds.top - pkg_top) < 1e-9
            and abs(bounds.bottom - pkg_bottom) < 1e-9
            and abs(px - pkg_px) < 1e-12
            and abs(py - pkg_py) < 1e-12
        )

        print("SAMPLE FILE   :", sample.name)
        print("WIDTH         :", src.width)
        print("HEIGHT        :", src.height)
        print("CRS           :", src.crs)
        print("BOUNDS        :", bounds)
        print("PIXEL_SIZE_X  :", px)
        print("PIXEL_SIZE_Y  :", py)
        print("MATCHES GRID  :", same_grid)

print("\n============================================================")
print("PROXY GRID CHECK COMPLETE")
print("============================================================")
```

```python
# ============================================================
# CELL 20 — BATCH ALIGN PROXY RASTERS TO AUTHORITATIVE GRID
# ============================================================

print("============================================================")
print("BATCH ALIGN PROXY RASTERS TO AUTHORITATIVE GRID")
print("============================================================")

from pathlib import Path
from rasterio.enums import Resampling
from pprint import pprint

SRC_BASE = Path(r"C:\Projects\Infer RozviDrought\rasters\corrected\proxies")
DST_BASE = Path(r"C:\Projects\Infer RozviDrought\data\aligned_grid")

VARIABLES = ["ndvi", "tws"]
SCENARIOS = ["historical", "ssp245", "ssp370", "ssp585"]

proxy_results = {}
proxy_failures = {}

for var in VARIABLES:
    proxy_results[var] = {}
    proxy_failures[var] = {}

    print("\n============================================================")
    print(f"VARIABLE: {var}")
    print("============================================================")

    for scenario in SCENARIOS:
        src_dir = SRC_BASE / var / scenario

        if scenario == "historical":
            dst_dir = DST_BASE / "historical" / var
        else:
            dst_dir = DST_BASE / "future" / scenario / var

        dst_dir.mkdir(parents=True, exist_ok=True)

        src_files = sorted(src_dir.glob("*.tif"))

        print("\n------------------------------------------------------------")
        print("SCENARIO         :", scenario)
        print("SOURCE DIRECTORY :", src_dir)
        print("OUTPUT DIRECTORY :", dst_dir)
        print("FILES FOUND      :", len(src_files))

        results = []
        failures = []

        if not src_files:
            failures.append({
                "variable": var,
                "scenario": scenario,
                "error": f"No source files found in {src_dir}",
            })
            proxy_results[var][scenario] = results
            proxy_failures[var][scenario] = failures
            print("No files found. Skipping.")
            continue

        for i, src_path in enumerate(src_files, start=1):
            dst_path = dst_dir / src_path.name

            try:
                result = align_raster_to_authoritative_grid(
                    src_path=src_path,
                    dst_path=dst_path,
                    resampling=Resampling.bilinear,
                )
                result["variable"] = var
                result["scenario"] = scenario
                result["file_index"] = i
                result["status"] = "ok"
                results.append(result)

                if i <= 2 or i == len(src_files):
                    print(f"[{i}/{len(src_files)}] OK : {src_path.name}")
                    print(
                        f"    min={result['min']:.6f} "
                        f"max={result['max']:.6f} "
                        f"mean={result['mean']:.6f} "
                        f"valid={result['valid_pixels']}"
                    )

            except Exception as e:
                failure = {
                    "variable": var,
                    "scenario": scenario,
                    "file_index": i,
                    "src_path": str(src_path),
                    "dst_path": str(dst_path),
                    "status": "failed",
                    "error": str(e),
                }
                failures.append(failure)

                print(f"[{i}/{len(src_files)}] FAILED : {src_path.name}")
                print(f"    error={e}")

        proxy_results[var][scenario] = results
        proxy_failures[var][scenario] = failures

        print("\nSCENARIO SUMMARY")
        print("------------------------------------------------------------")
        print("variable   :", var)
        print("scenario   :", scenario)
        print("total_files:", len(src_files))
        print("successes  :", len(results))
        print("failures   :", len(failures))

        if results:
            print("\nFIRST RESULT")
            print("------------------------------------------------------------")
            pprint(results[0])

        if failures:
            print("\nFIRST FAILURE")
            print("------------------------------------------------------------")
            pprint(failures[0])

print("\n============================================================")
print("PROXY ALIGNMENT COMPLETE")
print("============================================================")
```

```python
# ============================================================
# CELL 21 — BUILD SOURCE REGISTRY FROM ALIGNED RASTERS
# ============================================================

print("============================================================")
print("BUILD SOURCE REGISTRY")
print("============================================================")

from pathlib import Path
import re
import pandas as pd

ALIGNED_BASE = Path(r"C:\Projects\Infer RozviDrought\data\aligned_grid")
ARTIFACTS_DIR = Path(r"C:\Projects\Infer RozviDrought\data\artifacts")
ARTIFACTS_DIR.mkdir(parents=True, exist_ok=True)

HIST_VARIABLES = ["t2m", "d2m", "pet", "sm", "ndvi", "tws"]
FUTURE_VARIABLES = ["t2m", "d2m", "pet", "sm", "ndvi", "tws"]
SCENARIOS = ["historical", "ssp245", "ssp370", "ssp585"]

MONTH_PATTERN = re.compile(r"(\d{6})")

def extract_yyyymm(name: str):
    m = MONTH_PATTERN.search(name)
    return m.group(1) if m else None

rows = []

# ------------------------------------------------------------
# Historical
# ------------------------------------------------------------

for var in HIST_VARIABLES:
    var_dir = ALIGNED_BASE / "historical" / var
    files = sorted(var_dir.glob("*.tif"))

    for fp in files:
        yyyymm = extract_yyyymm(fp.name)
        if yyyymm is None:
            continue

        rows.append({
            "scenario": "historical",
            "yyyymm": yyyymm,
            "variable": var,
            "source_file": str(fp),
            "file_name": fp.name,
        })

# ------------------------------------------------------------
# Future
# ------------------------------------------------------------

for scenario in ["ssp245", "ssp370", "ssp585"]:
    for var in FUTURE_VARIABLES:
        var_dir = ALIGNED_BASE / "future" / scenario / var
        files = sorted(var_dir.glob("*.tif"))

        for fp in files:
            yyyymm = extract_yyyymm(fp.name)
            if yyyymm is None:
                continue

            rows.append({
                "scenario": scenario,
                "yyyymm": yyyymm,
                "variable": var,
                "source_file": str(fp),
                "file_name": fp.name,
            })

registry_df = pd.DataFrame(rows)

# ------------------------------------------------------------
# Basic validation
# ------------------------------------------------------------

registry_df = registry_df.sort_values(
    by=["scenario", "yyyymm", "variable"]
).reset_index(drop=True)

expected_pairs = []
for yyyymm in pd.read_parquet(
    r"C:\Projects\Infer RozviDrought\data\artifacts\master_month_index.parquet"
)["yyyymm"].astype(str):
    expected_pairs.append(("historical", yyyymm))

for scenario in ["ssp245", "ssp370", "ssp585"]:
    for yyyymm in pd.date_range("2026-01-01", "2050-12-01", freq="MS").strftime("%Y%m"):
        expected_pairs.append((scenario, yyyymm))

expected_df = pd.DataFrame(expected_pairs, columns=["scenario", "yyyymm"]).drop_duplicates()

coverage = (
    registry_df.groupby(["scenario", "yyyymm"])["variable"]
    .nunique()
    .reset_index(name="variables_found")
)

coverage = expected_df.merge(
    coverage,
    on=["scenario", "yyyymm"],
    how="left",
)

coverage["variables_found"] = coverage["variables_found"].fillna(0).astype(int)

# expected variable counts:
# historical has 6 variables but ndvi/tws start later naturally
# future scenarios should have 6 variables throughout
coverage["expected_variables"] = coverage["scenario"].apply(
    lambda s: 6 if s != "historical" else 6
)

# For historical, missing ndvi/tws early is expected; we keep registry factual.
# Still useful to detect months with zero files.
coverage["status"] = coverage["variables_found"].apply(
    lambda n: "missing_all" if n == 0 else "partial_or_complete"
)

# ------------------------------------------------------------
# Save
# ------------------------------------------------------------

registry_path = ARTIFACTS_DIR / "source_registry_198001_205012.parquet"
coverage_path = ARTIFACTS_DIR / "source_registry_coverage_198001_205012.parquet"

registry_df.to_parquet(registry_path, index=False)
coverage.to_parquet(coverage_path, index=False)

print("\nREGISTRY SUMMARY")
print("------------------------------------------------------------")
print("total_rows            :", len(registry_df))
print("unique_scenarios      :", registry_df['scenario'].nunique())
print("unique_months         :", registry_df['yyyymm'].nunique())
print("unique_variables      :", registry_df['variable'].nunique())

print("\nROWS PER SCENARIO")
print("------------------------------------------------------------")
print(registry_df.groupby("scenario").size())

print("\nCOVERAGE SUMMARY")
print("------------------------------------------------------------")
print("months_with_zero_files:", int((coverage["variables_found"] == 0).sum()))
print("future_months_not_6   :", int(
    coverage[coverage["scenario"] != "historical"]["variables_found"].ne(6).sum()
))

print("\nSAVED")
print("------------------------------------------------------------")
print(registry_path)
print(coverage_path)

print("\n============================================================")
print("SOURCE REGISTRY COMPLETE")
print("============================================================")
```

```python
# ============================================================
# CELL 22 — FIX COVERAGE LOGIC AND BUILD PIVOTED REGISTRY
# ============================================================

print("============================================================")
print("FIX COVERAGE LOGIC AND BUILD PIVOTED REGISTRY")
print("============================================================")

from pathlib import Path
import pandas as pd

ARTIFACTS_DIR = Path(r"C:\Projects\Infer RozviDrought\data\artifacts")
registry_path = ARTIFACTS_DIR / "source_registry_198001_205012.parquet"

registry_df = pd.read_parquet(registry_path)
registry_df["yyyymm"] = registry_df["yyyymm"].astype(str)

# ------------------------------------------------------------
# Correct expected month ranges
# ------------------------------------------------------------

historical_months = pd.date_range("1980-01-01", "2026-03-01", freq="MS").strftime("%Y%m")
future_months = pd.date_range("2026-01-01", "2050-12-01", freq="MS").strftime("%Y%m")

expected_pairs = []

for yyyymm in historical_months:
    expected_pairs.append(("historical", yyyymm))

for scenario in ["ssp245", "ssp370", "ssp585"]:
    for yyyymm in future_months:
        expected_pairs.append((scenario, yyyymm))

expected_df = pd.DataFrame(expected_pairs, columns=["scenario", "yyyymm"]).drop_duplicates()

coverage = (
    registry_df.groupby(["scenario", "yyyymm"])["variable"]
    .nunique()
    .reset_index(name="variables_found")
)

coverage = expected_df.merge(
    coverage,
    on=["scenario", "yyyymm"],
    how="left",
)

coverage["variables_found"] = coverage["variables_found"].fillna(0).astype(int)

# ------------------------------------------------------------
# Build pivoted registry: one row per scenario + yyyymm
# ------------------------------------------------------------

pivot_df = (
    registry_df.pivot_table(
        index=["scenario", "yyyymm"],
        columns="variable",
        values="source_file",
        aggfunc="first",
    )
    .reset_index()
)

pivot_df.columns.name = None

# rename to explicit path columns
rename_map = {
    "t2m": "t2m_path",
    "d2m": "d2m_path",
    "pet": "pet_path",
    "sm": "sm_path",
    "ndvi": "ndvi_path",
    "tws": "tws_path",
}
pivot_df = pivot_df.rename(columns=rename_map)

# join onto expected month scaffold
master_registry_df = expected_df.merge(
    pivot_df,
    on=["scenario", "yyyymm"],
    how="left",
)

# ------------------------------------------------------------
# Save
# ------------------------------------------------------------

coverage_path_fixed = ARTIFACTS_DIR / "source_registry_coverage_FIXED_198001_205012.parquet"
pivot_path = ARTIFACTS_DIR / "source_registry_pivoted_198001_205012.parquet"

coverage.to_parquet(coverage_path_fixed, index=False)
master_registry_df.to_parquet(pivot_path, index=False)

print("\nFIXED COVERAGE SUMMARY")
print("------------------------------------------------------------")
print("months_with_zero_files:", int((coverage['variables_found'] == 0).sum()))
print("future_months_not_6   :", int(
    coverage[coverage['scenario'] != 'historical']['variables_found'].ne(6).sum()
))

print("\nPIVOTED REGISTRY SUMMARY")
print("------------------------------------------------------------")
print("rows       :", len(master_registry_df))
print("columns    :", list(master_registry_df.columns))

print("\nSAVED")
print("------------------------------------------------------------")
print(coverage_path_fixed)
print(pivot_path)

print("\n============================================================")
print("FIXED COVERAGE AND PIVOT COMPLETE")
print("============================================================")
```

```python
# ============================================================
# CELL 23 — BUILD AUTHORITATIVE SPATIAL CELL INDEX
# ============================================================

print("============================================================")
print("BUILD AUTHORITATIVE SPATIAL CELL INDEX")
print("============================================================")

from pathlib import Path
import numpy as np
import pandas as pd
import rozvidrought_datasets.grid.spec as grid_spec

ARTIFACTS_DIR = Path(r"C:\Projects\Infer RozviDrought\data\artifacts")
ARTIFACTS_DIR.mkdir(parents=True, exist_ok=True)

WIDTH = grid_spec.WIDTH
HEIGHT = grid_spec.HEIGHT
LEFT = grid_spec.LEFT
TOP = grid_spec.TOP
PIXEL_SIZE_X = abs(grid_spec.PIXEL_SIZE_X)
PIXEL_SIZE_Y = abs(grid_spec.PIXEL_SIZE_Y)

rows = []

for row in range(HEIGHT):
    for col in range(WIDTH):
        lon = LEFT + (col + 0.5) * PIXEL_SIZE_X
        lat = TOP - (row + 0.5) * PIXEL_SIZE_Y
        pixel_id = row * WIDTH + col

        rows.append({
            "pixel_id": pixel_id,
            "row": row,
            "col": col,
            "lon": lon,
            "lat": lat,
        })

grid_index_df = pd.DataFrame(rows)

out_path = ARTIFACTS_DIR / "authoritative_grid_index.parquet"
grid_index_df.to_parquet(out_path, index=False)

print("\nGRID INDEX SUMMARY")
print("------------------------------------------------------------")
print("rows        :", len(grid_index_df))
print("width       :", WIDTH)
print("height      :", HEIGHT)
print("first_pixel :", grid_index_df.iloc[0].to_dict())
print("last_pixel  :", grid_index_df.iloc[-1].to_dict())

print("\nSaved:")
print(out_path)

print("\n============================================================")
print("AUTHORITATIVE SPATIAL CELL INDEX COMPLETE")
print("============================================================")
```

```python
# ============================================================
# CELL 24 — BUILD MASTER LONG-FORMAT DATASET
# One row per: scenario, yyyymm, pixel_id
# ============================================================

print("============================================================")
print("BUILD MASTER LONG-FORMAT DATASET")
print("============================================================")

from pathlib import Path
import numpy as np
import pandas as pd
import rasterio

ARTIFACTS_DIR = Path(r"C:\Projects\Infer RozviDrought\data\artifacts")
MASTER_DIR = Path(r"C:\Projects\Infer RozviDrought\data\master_inputs")
MASTER_DIR.mkdir(parents=True, exist_ok=True)

REGISTRY_PATH = ARTIFACTS_DIR / "source_registry_pivoted_198001_205012.parquet"
GRID_INDEX_PATH = ARTIFACTS_DIR / "authoritative_grid_index.parquet"

OUT_PATH = MASTER_DIR / "master_inputs_long_198001_205012.parquet"

VARIABLES = ["t2m", "d2m", "pet", "sm", "ndvi", "tws"]
PATH_COLUMNS = [f"{v}_path" for v in VARIABLES]
NODATA = -9999.0

registry_df = pd.read_parquet(REGISTRY_PATH)
grid_index_df = pd.read_parquet(GRID_INDEX_PATH)

registry_df["yyyymm"] = registry_df["yyyymm"].astype(str)

all_parts = []

def read_flat_raster(path):
    with rasterio.open(path) as src:
        arr = src.read(1).astype(np.float32)
    arr[arr == NODATA] = np.nan
    return arr.reshape(-1)

for i, rec in enumerate(registry_df.itertuples(index=False), start=1):
    month_df = grid_index_df.copy()
    month_df["scenario"] = rec.scenario
    month_df["yyyymm"] = rec.yyyymm

    for var in VARIABLES:
        path = getattr(rec, f"{var}_path", None)
        if pd.isna(path) or path is None:
            month_df[var] = np.nan
        else:
            month_df[var] = read_flat_raster(path)

    all_parts.append(month_df)

    if i <= 2 or i == len(registry_df):
        print(f"[{i}/{len(registry_df)}] OK : {rec.scenario} {rec.yyyymm}")

master_df = pd.concat(all_parts, ignore_index=True)

master_df.to_parquet(OUT_PATH, index=False)

print("\nMASTER DATASET SUMMARY")
print("------------------------------------------------------------")
print("rows         :", len(master_df))
print("columns      :", list(master_df.columns))
print("unique_months:", master_df[['scenario', 'yyyymm']].drop_duplicates().shape[0])
print("unique_cells :", master_df['pixel_id'].nunique())

print("\nSaved:")
print(OUT_PATH)

print("\n============================================================")
print("MASTER LONG-FORMAT DATASET COMPLETE")
print("============================================================")
```
